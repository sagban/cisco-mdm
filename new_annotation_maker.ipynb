```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
IN_P = os.path.join(os.getcwd(), "annotations_total_2649.csv")
OUT_P = os.path.join(os.getcwd(), "new_annotations")
```

```python
df = pd.read_csv(IN_P, header=None)
```

```python
df.rename(columns={0: "class", 1: "xmin", 2: "ymin", 3: "w", 4: "h", 5: "filename", 6: "width", 7: "height"}, inplace=True)
```

```python
df.head()
```

```python
my_groups = df[['class', 'filename']].groupby('filename')
```

```python
groupnames = list(my_groups.groups.keys())
```

```python
for i in groupnames:
    print("Filename {} classes {}".format(i, list(set(my_groups.get_group(i)['class'].values))))
    print("*"*30)
```

```python
filepath = os.path.join(OUT_P, "new_annotations.txt")

with open(filepath, 'w') as f:
    for i in groupnames:
        string = "{} ".format(i)
        for j in list(set(my_groups.get_group(i)['class'].values)):
            string = string + "{} ".format(j)
        f.write(string.rstrip(' ') + "\n")
        
```

# Making new labels

```python
lines = []
with open(os.path.join(OUT_P, "new_annotations.txt"), 'r') as f:
    lines = f.readlines()
```

```python
labels = ['rice', 'roti', 'dal', 'sabzi']

vecs = {labels[i]:np.eye(4)[i] for i in range(len(labels))}
print(vecs)
vecs['rice'].shape
```

```python
with open(os.path.join(OUT_P, "new_annotations_multihot.txt"), 'w') as f2:
    for i in lines:
        line = i.rstrip('\n')
        tokens = line.split(' ')
        filename, *labels = tokens
        multi_hot = np.zeros(shape=(4,))
        for j in labels:
            multi_hot = multi_hot + vecs[j]
        string = filename
        for k in multi_hot:
            string = string + " " + str(k)
        f2.write(string + '\n')    
```

# One Hot

```python
with open(os.path.join(OUT_P, "new_annotations_multihot.txt"), 'r') as f3:
    lines = f3.readlines()
    for i in lines:
        line = i.rstrip('\n')
        tokens = line.split(' ')
        filename, *labels = tokens
        multi_hot = np.array(labels)
        print("filename {} multihot: {}".format(filename, multi_hot))
        print("*"*30)
```

```python
def bin2dec(vec):
    result = 0.0
    for i in range(len(vec)):
        result += vec[i] * (2**(len(vec)-i-1))
    return result    
```

```python
with open(os.path.join(OUT_P, "new_annotations_onehot_1.txt"), 'w') as f_w:
    with open(os.path.join(OUT_P, "new_annotations_multihot.txt"), 'r') as f:
        lines = f.readlines()
        for i in lines:
            line = i.rstrip('\n')
            tokens = line.split(' ')
            filename, *labels = tokens
            multi_hot = np.array(labels, dtype=np.float32)
            print(type(multi_hot[0]))
            #print("filename {} multihot: {}".format(filename, multi_hot))
            #print("*"*30)
            dec = bin2dec(multi_hot)
            f_w.write('{} {}\n'.format(filename, str(dec)))
        
```
